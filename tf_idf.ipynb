```python
import math
```

```python
import re # elimate - expression marks
```

```python
import collections # to use Counter
```

```python
import pandas as pd
```

```python
query = input("Enter the Search.. ")
```

```python
corpus = [
    "Machine learning is transforming the way computers learn from data.",
    "Learning algorithms analyze massive datasets to identify patterns that humans cannot easily detect.",
    "Data is the fuel of machine learning without quality data even the most powerful algorithms fail.",
    "Deep learning a subset of machine learning uses neural networks to process data at multiple layers.",
    "Neural networks are inspired by the human brain and can learn complex patterns from raw data.",
    "Algorithms like gradient descent optimize neural networks by minimizing error across training data.",
    "Training a deep learning model requires enormous computational power and carefully labeled data.",
]
```

```python
all_terms = []
doc_data = []
local_terms = [] # document level - term frequency

for doc in corpus:
  doc = re.sub(r'[^\w\s]','',doc.lower())
  doc = doc.split()
  doc_data.append(doc)

  local_terms.append(collections.Counter(doc))

  all_terms.extend(doc)
```

```python
global_terms = collections.Counter(all_terms) # global - term frequency
```

```python
idf_scores = {} # store IDF scores of all terms
no_of_docs = len(doc_data)
for word in set(all_terms):
  count_of_term = 0
  for idx, doc in enumerate(local_terms):
    if word in doc:
      count_of_term +=1
  idf_scores[word]= math.log(no_of_docs/count_of_term)
```

```python
tfidf_matrix = [] # creating a matrix
for id, line in enumerate(local_terms):
  t = line.total()
  doc_tfidf= {}
  for word in list(global_terms.keys()):
    f = line.get(word, 0)
    tf = f/t
    doc_tfidf[word] = tf*idf_scores[word]
  tfidf_matrix.append(doc_tfidf)
```

```python
print(f"{'':12}", end="")
for term in list(global_terms.keys())[:6]:             # show first 6 terms for readability
    print(f"{term:>12}", end="")
print()

for i, doc_vec in enumerate(tfidf_matrix):
    print(f"Doc {i+1:>6}  ", end="")
    for term in list(global_terms.keys())[:6]:
        print(f"{doc_vec[term]:>12.4f}", end="")
    print()
```
