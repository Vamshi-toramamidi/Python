```python
import re
import collections
import nltk
from nltk.corpus import stopwords
import tensorflow as tf

# Ensure NLTK stopwords are downloaded
try:
    nltk.data.find('corpora/stopwords')
except LookupError:
    nltk.download('stopwords')

input_sentence = "Python is a versatile programming language, python is SuPer Crazy. pYthon is easy to learn"

# 1. Lowercase + strip punctuation in one pass
cleaned = re.sub(r'[^\w\s]', '', input_sentence.lower())
tokens = cleaned.split()

# 2. Remove stopwords (set lookup is O(1))
stop_words = set(stopwords.words('english'))
filtered = [w for w in tokens if w not in stop_words]

# 3. Count words
word_counts = collections.Counter(filtered)

# 4. Build vocabulary: word -> index (standard convention)
vocabulary = {word: idx for idx, word in enumerate(word_counts)}

# 5. Build matrix rows directly
matrix = [[vocabulary[word], word, count] for word, count in word_counts.items()]

print("\n--- Bag-of-Words Matrix (Index, Word, Count) ---")
for row in matrix:
    print(row)

print("\n--- Vocabulary (Word: Index) ---")
print(vocabulary)

# 6. TensorFlow tensor — built directly, no second loop needed
tensor = tf.constant([[vocabulary[word], count] for word, count in word_counts.items()], dtype=tf.int32)
print("\n--- TensorFlow Tensor (Index, Count) ---")
print(tensor)

# 7. Presence indicators — use a set for O(1) lookup, fix the indentation bug
test_sentence = set("python is versatile".lower().split())

# Sorted by index so output order is deterministic
sorted_vocab = sorted(vocabulary.items(), key=lambda x: x[1])  # [(word, idx), ...]
indices       = [idx  for _, idx  in sorted_vocab]
presence      = [1 if word in test_sentence else 0 for word, _ in sorted_vocab]

print("\n--- Test Sentence Presence Indicators ---")
print([indices, presence])
```

**TF**-**IDF**

```python
import math
import collections
import re
```

```python
import pandas as pd
```

```python
corpus = [
    "Machine learning is transforming the way computers learn from data.",
    "Learning algorithms analyze massive datasets to identify patterns that humans cannot easily detect.",
    "Data is the fuel of machine learning — without quality data, even the most powerful algorithms fail.",
    "Deep learning, a subset of machine learning, uses neural networks to process data at multiple layers.",
    "Neural networks are inspired by the human brain and can learn complex patterns from raw data.",
    "Algorithms like gradient descent optimize neural networks by minimizing error across training data.",
    "Training a deep learning model requires enormous computational power and carefully labeled data.",
]
```

```python
all_terms = []
row_data = [] # row_data - term - each document

for line in corpus:
  cleaned = re.sub(r'[^\w\s]', '', line.lower())
  temp = cleaned.split()
  row_data.append(collections.Counter(temp))
  all_terms.extend(temp)
vocabulary = dict(collections.Counter(all_terms))

dictionary = dict(sorted(vocabulary.items(), key=lambda item: item[1], reverse=True)) # dictionary - all term - all documents
```

```python
val = input("Enter the element: ")

doc_no = [id for id, doc in enumerate(row_data)]
searched = dict.fromkeys(doc_no)
print(searched)

count = 0
for i, counter_ in enumerate(row_data):
  f = counter_.get(val, 0)
  if f != 0:
    count+=1

if count == 0:
  print(f"the term '{val}' is not present in any document")

else:
  idf = math.log(len(row_data)/count) # Inverse Document Frequency
  print (f"\ninverse document frequency for '{val}' : {idf}")
  for i, counter_ in enumerate(row_data):
    f = counter_.get(val, 0)
    t = int(counter_.total())
    tf = f/t # Term Frequency
    tf_idf = tf*idf # TF - IDF
    print(f"doc{i+1}: {tf}")
    searched[i] = tf_idf

print(searched)
searched = dict(sorted(searched.items(), key=lambda x: x[1], reverse=True))
print(f"\nRanked Documents: {searched}")
```
